```python
df = raw_data.copy()
```

```python
# useless columns
df.drop(['id', 'member_id', 
         'zip_code', 'url', 
         'emp_title',
        'desc', 'addr_state',
        'out_prncp_inv', 'total_pymnt_inv',
        'policy_code', 'application_type'], axis=1, inplace=True)
```

```python
df.head()
```

## Data Cleaning

```python
# remove percent signs and convert to float
for column in ['int_rate', 'revol_util']:
    df[column] = df[column].str.replace('%', '').astype('f')
```

```python
df['term'] = df['term'].str.replace(' months', '').astype('i')
```

```python
%time missing_vals = df.apply(lambda x: (sum(x.isnull().values), sum(x.isnull().values)/ len(df)), axis = 0) # For columns
```

```python
missing_vals.sort_values(ascending=False)[:36]
```

```python
df.dropna(axis=1, how='any', inplace=True)
```

```python
# remove date columns
df.drop(['issue_d', 'earliest_cr_line'], axis=1, inplace=True)
```

```python
df.info()
```

## Model Creation

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
# drop all categorical variables
X = df.drop(['grade', 'sub_grade', 'emp_length', 'home_ownership', 'verification_status',
            'loan_status', 'pymnt_plan', 'purpose', 'initial_list_status'], axis=1)
```

```python
y = df.grade
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
cross_val_score(rfc, X, y, cv=5)
```
